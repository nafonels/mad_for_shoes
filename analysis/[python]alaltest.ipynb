```python
# 날짜변환을 위한 예제 처리
# dist_tw_cmt_data['asDate']=pd.to_datetime(dist_tw_cmt_data['post_date'], format='%Y-%m-%d')
# %c Mon Sep 30 07:06:05 2013
# '%a %b %d %H:%M:%S %f %Y'
nddate = datetime.datetime.strptime(dist_tw_post_data['post_date'][1], '%a %b %d %H:%M:%S +0000 %Y').strftime('%Y-%m-%d')
nddate
```

```python
nbdate=pd.to_datetime(nddate,format='%Y-%m-%d')
nbdate
```

```python
#날짜를 변환한다.(dist2)
for i in range(len(dist_tw_post_data)):
    nddate = datetime.datetime.strptime(dist_tw_post_data['post_date'][i], '%a %b %d %H:%M:%S +0000 %Y').strftime('%Y-%m-%d')
    dist_tw_post_data.loc[dist_tw_cmt_data.index==i,'asDate'] = nddate
```
